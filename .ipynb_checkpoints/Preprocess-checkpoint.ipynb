# Imports

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

import os
import glob
import re

np.random.seed(42)

pd.options.display.max_columns = None
```

```python
path = r'Indexes'
df = pd.DataFrame(columns=['Date'])
df['Date'] = pd.date_range('01-01-2000','01-01-2021',freq='D')

for filename in glob.glob(os.path.join(path, '*.csv')):
    col = re.search(r".*\\(.+).csv",filename)[1]

    if col in ['DowJones','Gold','Ibovespa','Nasdaq','S&P','SMAL11-ETF','Vix','IVBX2','USDBRL']:
        aux = pd.read_csv(filename,sep=',')
        aux.columns = ['Date','Open','High','Low','Close',col+'_Close','Volume']
    elif col == 'selic':
        aux = pd.read_csv(filename,sep=';')
        aux = aux.iloc[:,:2]
        aux.columns = ['Date',col+'_Close']
    else:
        aux = pd.read_csv(filename,sep=';')
        aux.columns = ['Date',col +'_Close','Open','High','Low','Vol','Variation','Junk']
        #aux.replace('.','', inplace =True, regex=True)
        
        
    aux['Date'] = pd.to_datetime(aux['Date'])
    if type(aux[col+'_Close'].unique()[1]) is str: 
        aux[col+'_Close'] = aux[col+'_Close'].str.replace('.','',regex=True).replace(',','.',regex=True).astype(float)
    
    df = df.merge(aux[['Date',col+'_Close']], on='Date', how='outer')
    
    df.index = df['Date']
    df.drop(columns='Date',inplace=True)
```

```python
df.head()
```

```python
log_ret = np.log(df/df.shift(1))
```

```python
log_ret
```

```python
num_ports = 6000
all_weights = np.zeros((num_ports, len(df.columns)))
ret_arr = np.zeros(num_ports)
vol_arr = np.zeros(num_ports)
sharpe_arr = np.zeros(num_ports)

for x in range(num_ports):
    # Weights
    weights = np.array(np.random.random(len(df.columns)))
    weights = weights/np.sum(weights)
    
    # Save weights
    all_weights[x,:] = weights
    
    # Expected return
    ret_arr[x] = np.sum( (log_ret.mean() * weights * 252))
    
    # Expected volatility
    vol_arr[x] = np.sqrt(np.dot(weights.T, np.dot(log_ret.cov()*252, weights)))
    
    # Sharpe Ratio
    sharpe_arr[x] = ret_arr[x]/vol_arr[x]
```

```python
plt.figure(figsize=(12,8))
plt.scatter(vol_arr, ret_arr, c=sharpe_arr, cmap='viridis')
plt.colorbar(label='Sharpe Ratio')
plt.xlabel('Volatility')
plt.ylabel('Return')
#plt.scatter(max_sr_vol, max_sr_ret,c='red', s=50) # red dot
plt.show()
```

```python
def get_ret_vol_sr(weights):
    weights = np.array(weights)
    ret = np.sum(log_ret.mean() * weights) * 252
    vol = np.sqrt(np.dot(weights.T, np.dot(log_ret.cov()*252, weights)))
    sr = ret/vol
    return np.array([ret, vol, sr])

def neg_sharpe(weights):
# the number 2 is the sharpe ratio index from the get_ret_vol_sr
    return get_ret_vol_sr(weights)[2] * -1

def check_sum(weights):
    #return 0 if sum of the weights is 1
    return np.sum(weights)-1
```
