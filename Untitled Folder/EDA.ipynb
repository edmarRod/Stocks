# Imports

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

np.random.seed(42)

pd.options.display.max_columns = None
```

# Get data

```python
df = pd.read_csv(r"../data/processed/processed.csv")
df['Date'] = pd.to_datetime(df['Date'])
df.index=df['Date']
df.drop(columns=['Date'], inplace=True)
display(df)
```

# Compare main indexes

## Timeline

```python
df_main = df[['Ibovespa_Close','S&P_Close','USDBRL_Close','selic_Close','Vix_Close','Nasdaq_Close','Gold_Close','DowJones_Close']]
df_main.plot(subplots=True, figsize=(20,10))
```

## Correlations

```python
df_main.corr()
```

# Compare Ibovespa indexes

```python
df_ibov = df[df.columns.difference(['S&P_Close','USDBRL_Close','selic_Close','Vix_Close',
                                         'Nasdaq_Close','DowJones_Close','Gold_Close'])]
display(df_ibov.head())
```

```python
sns.heatmap(df_ibov.corr())
```

```python
df_ibov.corr()
```

# Calculate returns
- Selic is a special case since the data is annual rate
- Current returns are daily

```python
df_returns = 1-df/df.shift(-1)
df_returns['selic_Close'] = 1-df['selic_Close']**(1/365)
```

```python
display(df_returns.head())
```

# Compare returns of main indexes

```python
df_main = df_returns[['Ibovespa_Close','S&P_Close','USDBRL_Close','selic_Close','Vix_Close','Nasdaq_Close','Gold_Close','DowJones_Close']]
df_main.plot(subplots=True, figsize=(20,10))
```

```python
df_main.corr()
```

# Compare returns of Ibovesepa indexes

```python
df_ibov = df_returns[df_returns.columns.difference(['S&P_Close','USDBRL_Close','selic_Close','Vix_Close',
                                         'Nasdaq_Close','DowJones_Close','Gold_Close'])]
display(df_ibov.head())
```

```python
sns.heatmap(df_ibov.corr())
```

```python
df_ibov.corr()
```
