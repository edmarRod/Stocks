# Imports

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

np.random.seed(42)

pd.options.display.max_columns = None
```

# Get data

```python
df = pd.read_csv(r"../data/processed/processed.csv")
df['Date'] = pd.to_datetime(df['Date'])
df.index=df['Date']
df.drop(columns=['Date'], inplace=True)
display(df)
```

# Compare everything

```python
df.plot(subplots=True, layout=(5,5), figsize=(20,10))
plt.show()
```

# Compare main indexes

## Timeline

```python
df_main = df[['Ibovespa_Close','S&P_Close','USDBRL_Close','selic_Close','Vix_Close','Nasdaq_Close','Gold_Close','DowJones_Close']]
df_main.plot(subplots=True, figsize=(20,10))
plt.show()
```

## Correlations

```python
df_main.corr()
```

# Compare Ibovespa indexes

```python
df_ibov = df[df.columns.difference(['S&P_Close','USDBRL_Close','selic_Close','Vix_Close',
                                         'Nasdaq_Close','DowJones_Close','Gold_Close'])]
display(df_ibov.head())
```

```python
sns.heatmap(df_ibov.corr())
```

```python
df_ibov.corr()
```

# Calculate returns
- Selic is a special case since the data is annual rate
- Current returns are daily

```python
df_returns = df/df.shift(1)
df_returns['selic_Close'] = df['selic_Close']**(1/365)
```

```python
display(df_returns.head())
```

# Compare returns of main indexes

```python
df_main = df_returns[['Ibovespa_Close','S&P_Close','USDBRL_Close','selic_Close','Vix_Close','Nasdaq_Close','Gold_Close','DowJones_Close']]
df_main.plot(subplots=True, figsize=(20,10))
plt.show()
```

```python
df_main.corr()
```

# Compare returns of Ibovesepa indexes

```python
df_ibov = df_returns[df_returns.columns.difference(['S&P_Close','USDBRL_Close','selic_Close','Vix_Close',
                                         'Nasdaq_Close','DowJones_Close','Gold_Close'])]
display(df_ibov.head())
```

```python
sns.heatmap(df_ibov.corr())
```

```python
df_ibov.corr()
```

# Risk assessment
- Risk = volatility measured by sd

## All time

```python
df_risk = df_returns.describe().T.sort_values(['mean'],ascending=False)
display(df_risk)
```

## Pre 2020 and covid

```python
df_risk = df_returns.loc[df_returns.index < '2020-01-01'].describe().T.sort_values(['mean'],ascending=False)
display(df_risk)
```

## Last 10 years

```python
df_risk = df_returns.loc[df_returns.index > '2010-01-01'].describe().T.sort_values(['mean'],ascending=False)
display(df_risk)
```

## Last year

```python
df_risk = df_returns.loc[df_returns.index > '2020-01-01'].describe().T.sort_values(['mean'],ascending=False)
display(df_risk)
```

# Monthly returns

```python
df_returns = df.groupby(df.index.strftime('%Y-%m')).max()/df.groupby(df.index.strftime('%Y-%m')).max().shift(1)
df_returns.describe().T.sort_values(['mean'],ascending=False)
```

# Annualized returns

```python
df_returns = df.groupby(df.index.year).max()/df.groupby(df.index.year).max().shift(1)
df_returns.describe().T.sort_values(['mean'],ascending=False)
```
